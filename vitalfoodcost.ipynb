```python

#plot
import matplotlib.pyplot as plt

#import math
import math

#import seaborn
import seaborn as sns; sns.set_theme()
```

```python
#read the .csv and display the info
df_food = pd.read_csv ('/kaggle/input/food-prices/Food Prices.csv')
df_food.info()
df_food
```

```python
#Display the Country
UnqCountry = df_food['Country '].unique()
print ("The countries we analyzed are ", UnqCountry)
```

```python
#Display the Year
UnqYear = df_food['Year '].unique()
print ("The years we analyzed are ", UnqYear)
```

```python
#Display the Food Item
UnqItem = df_food['Food Item '].unique()
print ("The Food Item we analyzed are ", UnqItem)
```

```python
#Average of price
df_avgPrice=df_food.groupby(['Country ','Year ','Food Item '])['Average Price '].mean().round()

df=pd.DataFrame(df_avgPrice)
print(df)
```

```python
# Plotting a box plot to visualize the distribution of average prices for different food items
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_food, x='Food Item ', y='Average Price ',hue='Year ', palette='Set2')
plt.xlabel('Food Item')
plt.ylabel('Average Price (AUD)')
plt.title('Distribution of Average Prices for Different Food Items')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```
